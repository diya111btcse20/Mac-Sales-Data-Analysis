```python
#Create a new Column named 'Month' 
sales_data['Month'] = sales_data['Order Date'].str[0:2]
sales_data.head()
```

```python
filt = sales_data['Quantity Ordered'] == 'Quantity Ordered'
new_df = sales_data.loc[-filt]
```

```python
new_df.shape
```

```python
new_df.loc[filt]
```

```python
new_df
```

```python
new_df['Quantity Ordered'] = new_df['Quantity Ordered'].astype('int32')
```

```python
new_df['Price Each'] = new_df['Price Each'].astype('float64')
```

```python
new_df.info()
```

```python
new_df['Sales'] = (new_df['Quantity Ordered'] * new_df['Price Each'])
```

```python
new_df.info()
```

```python
new_df.head()
```

```python
ans_df = new_df.groupby('Month').sum()
ans_df
```

```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')


month = ans_df.index
sales = ans_df['Sales']



plt.plot(month, sales, color="#444444", linewidth=2, marker='o')

plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales in USD')

plt.show()
```

```python
new_df.head()
```

```python
#Create new Column 'City'
new_df['City'] = new_df['Purchase Address'].apply(lambda x: x.split(',')[1])
new_df.head()
```

```python
new_df
```

```python
city_grp = new_df.groupby(['City'])
```

```python
ans_df2 = city_grp['Sales'].sum()
ans_df2.rename('Units')
```

```python
plt.style.use('seaborn')

plt.plot(ans_df2, color='#444444', marker='o', linewidth=2)

plt.title('City wise \n Sales')
plt.xlabel('City')
plt.xticks(rotation='45', size=13)
plt.yticks(size=13)
plt.ylabel('Sales in USD')

plt.show()
```

```python
new_df.head()
```

```python
ans_df3 = new_df.groupby(['Product']).sum()
ans_df3 = ans_df3.sort_values(by='Quantity Ordered')
```

```python
ans_df3
```

```python
products = ans_df3.index
quantity = ans_df3['Quantity Ordered']

plt.style.use('seaborn')

plt.barh(products, quantity)
plt.xlabel('Units')
plt.xticks(size=15)
plt.yticks(size=15)

plt.show()
```

```python
ans_df3 = ans_df3.sort_values(by='Sales')
```
